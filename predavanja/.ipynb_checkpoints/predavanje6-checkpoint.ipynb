```python
# s[:n] - prvih n
# s[:-n] - brez zadnjih n
# s[n:] - brez prvih n
# s[-n:] - zadnjih n
print(imena)
print(imena[3])      # cetrti
print(imena[:2])     # prva dva
print(imena[:-2])    # brez zadnjih dveh
print(imena[2:])     # brez prvih dveh
print(imena[-2:])    # zadnja dva
```

```python
from itertools import *
```

```python
list(pairwise(imena))
```

```python
for prejsnji, trenutni in pairwise(imena):
    print(prejsnji, trenutni)

# sestavi seznam zaporednih parov
```

```python
list(enumerate(imena))
```

```python
for i, ime in enumerate(imena):
    print(ime, imena[i - 1])
```

```python
#matematik Erdoš

#sodo stevilo delimo z 2
#liho stevilo pristejemo 1 in mnozimo s 3
#ce pridemo do 1, zmagamo
#ce se zaciklamo v neskoncno, zgubimo

n = 14

while n != 1:
    print(n)
    if n % 2 == 0:
        n = n // 2
    else:
        n = 3 * n + 1
```

```python
#izpis 10 stevil med 20 in 30
n = 20
while n < 30:
    print(n)
    n += 1
```
